```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {'Nombre': ['Ana', 'Juan', 'Pedro', 'María'],
        'Edad': [25, 30, 35, 28],
        'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']}
df = pd.DataFrame(data)

# Mostrar el DataFrame
display(df)

# Crear un histograma de las edades
plt.hist(df['Edad'])
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.show()

# Crear un gráfico de barras de las ciudades
plt.bar(df['Ciudad'], df['Edad'])
plt.xlabel('Ciudad')
plt.ylabel('Edad')
plt.title('Gráfico de Barras de Edades')
```
